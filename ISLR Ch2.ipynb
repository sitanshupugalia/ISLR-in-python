```python
# Question 10
from sklearn.datasets import load_boston
boston = load_boston()
df = pd.DataFrame(boston.data, columns = boston.feature_names) # to convert the index to the column names.
df['Price'] = boston.target # the Price column is available in other attribute called target.
df # to know more about target, check https://medium.com/@haydar_ai/learning-data-science-day-9-linear-regression-on-boston-housing-dataset-cd62a80775ef
```

```python
# 10(b)
g = sns.PairGrid(df)
g.map_upper(plt.scatter, s = 3, color = 'red')
g.map_diag(plt.hist)
g.map_lower(plt.scatter, s = 3, color = 'red') 
g.fig.set_size_inches(12, 12)
```

```python
plt.scatter(df['RM'], df['Price'])
plt.xlabel('RM')
plt.ylabel('Price')
# Output represents: RM and target has positive linear relationship. This is expected as RM is the number of rooms (more space, higher price)
```

```python
plt.scatter(df['LSTAT'], df['Price'])
plt.xlabel('LSTAT')
plt.ylabel('Price')
# Output represents: LSTAT and target has negative non-linear relationship. This is expected as LSTAT is the percent of lower status people (lower status, lower incomes, cheaper houses)
```

```python
plt.scatter(df['LSTAT'], df['RM'])
plt.xlabel('LSTAT')
plt.ylabel('RM')
# Output represents: LSTAT and target has negative non-linear relationship. It makes sense since people with less money (higher LSTAT) can't afford bigger houses (high RM)
```
